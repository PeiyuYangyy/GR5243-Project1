## Train and Test Split

```python
X = housing_price_data[feature_name]
y = housing_price_data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

# Model Fitting and hyperparameter Tuning

## Multi-Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X_train, y_train)
for feature, coeff in zip(list(X_train.columns), mlr.coef_):
    if coeff > 0:
        positive_fea = feature
        positive_co = coeff
        print(f'{positive_fea} has a positive relationship {positive_co:.2f}')
    else:
        print(f' {feature} has the coefficient {coeff:.2f}')
```

```python
# mlr_training_r2 = mlr.score(X_train, y_train)
# print(f'mlr training set R^2: {mlr_training_r2:.2f}')
```

```python
# X_test.isna().sum(), y_test.isna().sum()
```

```python
# mlr_test_r2 = mlr.score(X_test, y_test)
# print(f'mlr test R^2: {mlr_test_r2:.2f}')
```

```python
from sklearn.model_selection import cross_val_score
mlr_cv_scores = cross_val_score(LinearRegression(), X_train, y_train, cv=10)
mlr_cv_mean = mlr_cv_scores.mean()

mlr_cv_2std = 2*mlr_cv_scores.std()

print(f'mlr mean cv r2: {mlr_cv_mean:.2f} +- {mlr_cv_2std:.2f}')
```

## Decision Tree Regressor Model

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

params_dtr = {'max_depth': [1, 2, 3, 5, 10]}

dtr_gscv = GridSearchCV(DecisionTreeRegressor(random_state=123), params_dtr, cv=10, refit=True, n_jobs=-1).fit(X_train, y_train)

print(f'dtr best hyperparams      : {dtr_gscv.best_params_}')
print(f'dtr best mean cv accuracy : {dtr_gscv.best_score_:.3f}')
```

```python
# dtr = DecisionTreeRegressor(max_depth=5)
# dtr.fit(X_train,y_train)
# dtr_train_r2 = dtr.score(X_train,y_train)
# print(f'dtr train R^2: {dtr_train_r2:.2f}')
```

```python
# dtr_test_r2 = dtr.score(X_test, y_test)
# print(f'dtr test R^2: {dtr_test_r2:.2f}')
```

```python
dtr_cv_scores = cross_val_score(DecisionTreeRegressor(max_depth=5, random_state=123), X_train, y_train, cv=10)
dtr_cv_mean = dtr_cv_scores.mean()

dtr_cv_2std = 2*dtr_cv_scores.std()

print(f'dtr mean cv r2: {dtr_cv_mean:.3f} +- {dtr_cv_2std:.3f}')
```

## Gradient Boosting Regressor Model

```python
from sklearn.ensemble import GradientBoostingRegressor

params_gbr = {'n_estimators': [10, 50, 100, 200, 500], 'max_depth': [1, 2, 3, 5, 10]}

gbr_gscv = GridSearchCV(GradientBoostingRegressor(random_state=123), params_gbr, cv=10 ,refit=True, n_jobs=-1).fit(X_train, y_train)

print(f'gbr best hyperparams      : {gbr_gscv.best_params_}')
print(f'gbr best mean cv accuracy : {gbr_gscv.best_score_:.3f}')
```

```python
gbr_cv_scores = cross_val_score(GradientBoostingRegressor(n_estimators=50, max_depth=3, random_state=123), X_train, y_train, cv=10, n_jobs=-1)
gbr_cv_mean = gbr_cv_scores.mean()
gbr_cv_2std = 2*gbr_cv_scores.std()
print(f'gbr mean cv r2: {gbr_cv_mean:.3f} +- {gbr_cv_2std:.3f}')
```

## XGBoost Regressor Model

```python
from xgboost import XGBRegressor

params_xgbr = {'n_estimators': [10, 50, 100, 200, 500], 'max_depth': [1, 2, 3, 5, 10], 'learning_rate': [0.01,0.1,0.5,1.0]}

xgbr_gscv = GridSearchCV(XGBRegressor(random_state=123), params_xgbr, cv=10 ,refit=True, n_jobs=-1).fit(X_train, y_train)

print(f'xgbr best hyperparams      : {xgbr_gscv.best_params_}')
print(f'xgbr best mean cv accuracy : {xgbr_gscv.best_score_:.3f}')
```

```python
xgbr_cv_scores = cross_val_score(XGBRegressor(n_estimators=50, learning_rate=0.1, max_depth=3, random_state=123),X_train, y_train, cv=10, n_jobs=-1)
xgbr_cv_mean = xgbr_cv_scores.mean()
xgbr_cv_2std = 2*xgbr_cv_scores.std()
print(f'xgbr mean cv r2: {xgbr_cv_mean:.3f} +- {xgbr_cv_2std:.3f}')
```
